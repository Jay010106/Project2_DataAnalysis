```python
import pandas as pd

# Load the local dataset
data = pd.read_csv('export.csv')

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(data.head())

```

```python
# Check for missing values in the dataset
print("Missing values in each column:")
print(data.isnull().sum())
```

```python
# Fill missing values with 'Unknown'
data.fillna('Unknown', inplace=True)

# Verify missing values are filled
print("Missing values after cleaning:")
print(data.isnull().sum())
```

```python
# Count Nobel Laureates by country
country_counts = data['bornCountry'].value_counts()

# Plot the Top 10 countries
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
country_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Most Nobel Laureates')
plt.xlabel('Country')
plt.ylabel('Number of Laureates')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count Nobel Laureates by gender
gender_counts = data['gender'].value_counts()

# Plot the gender distribution
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.title('Gender Distribution of Nobel Laureates')
plt.ylabel('')
plt.show()
```

```python
# Count Nobel Prizes by category
category_counts = data['category'].value_counts()

# Plot the prize distribution by category
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='orange')
plt.title('Number of Nobel Prizes by Category')
plt.xlabel('Category')
plt.ylabel('Number of Prizes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count Nobel Prizes awarded per year
year_counts = data['year'].value_counts().sort_index()

# Plot the trend of Nobel Prizes over the years
plt.figure(figsize=(12, 6))
year_counts.plot(kind='line', color='green')
plt.title('Number of Nobel Prizes Awarded Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Prizes')
plt.grid()
plt.show()
```

## Insights:
1. The United States and several European countries dominate the list of Nobel Laureates.
2. The Physics, Chemistry, and Medicine categories have awarded the most prizes.
3. There is a significant gender imbalance, with most prizes awarded to male recipients.
4. Nobel Prizes have been awarded consistently over the years, with some periods of rapid increase.
